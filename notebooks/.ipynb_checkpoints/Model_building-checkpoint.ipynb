#### Loading The DataSet & Data Understanding

```python
sales_data = pd.read_csv(r'C:\Users\amank\OneDrive\Desktop\Car_Forcasting\data\processed_data\sales_data_clean.csv')
```

```python
data = sales_data[['INVOICEDAT']]
```

```python
# Ensure the index is a datetime type and prepare data for Prophet
data = sales_data.INVOICEDAT.value_counts().reset_index().sort_values(by='INVOICEDAT').reset_index(drop=True)
data = data.set_index('INVOICEDAT')
data.index = pd.to_datetime(data.index)

# Convert the data to monthly frequency and sum values within each month
data = data.resample('M').sum()
data.head()
```

#### Checking the data is stationary or not

```python
#Determing rolling statistics
rolmean = data['count'].rolling (window=12).mean() 
rolstd = data['count'].rolling (window=12).std()
#Plot rolling statistics:
orig = plt.plot(data['count'], color='blue', label='Original')
mean = plt.plot(rolmean, color='red', label='Rolling Mean')
std = plt.plot(rolstd, color='black', label = 'Rolling Std') 
plt.title('Rolling Mean & Standard Deviation')
plt.xticks(ticks=data.index)
plt.legend()
plt.show()
```

#### Train-test Split

```python
# Split the data into training and test sets (for evaluation purposes)
train_size = int(len(data) * 0.8)
train_data, test_data = data[:train_size], data[train_size:]
```

#### Model Building

- **AR -** Auto Regressive 

- **MA -** Moving Average

- **ARMA -** Auto Regressive + Moving Avere\age

- **ARIMA -** Auto Regressive + Intregation + Moving Average

- **SARIMA -** Auto Regressive + Intregation + Moving Average + Seasonality

```python
from statsmodels.tsa.stattools import acf, pacf
import statsmodels.api as sm

# Calculate the maximum allowable lag
max_lags = data['count'].shape[0] // 2 - 1  # Adjust to be less than half the sample size

# Plot ACF and PACF with the limited lag
fig_acf = sm.graphics.tsa.plot_acf(data['count'], lags=max_lags)
fig_pacf = sm.graphics.tsa.plot_pacf(data['count'], lags=max_lags)
```

- **ARMA Model**

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import r2_score

# Fit the ARMA model 
model = ARIMA(train_data['count'], order=(6, 0, 6)) #(p, d, q)  (AR, I, MA)
arima_model = model.fit()

# Get predictions on the test data
forecast = arima_model.get_forecast(steps=len(test_data))
predictions = forecast.predicted_mean

# Create an index for predictions that matches the test data index
predictions_index = test_data.index

# Set the index of predictions
predictions = pd.Series(predictions, index=predictions_index)

# Calculate error metrics
mae = r2_score(test_data['count'], predictions)
mse = r2_score(test_data['count'], predictions)

print(f"R Squred: {mae}")
print(f"R Squred: {mse}")

# Optional: Plot the predictions vs actual values
plt.figure(figsize=(10, 6))
plt.plot(train_data.index, train_data['count'], label='Train', marker='o')
plt.plot(test_data.index, test_data['count'], label='Test', marker='o')
plt.plot(predictions.index, predictions, label='Predictions', color='g', marker='o')
plt.xlabel('Date')
plt.ylabel('Count')
plt.title('ARIMA Model Predictions')
plt.axvline(test_data.index[0], color='r', linestyle='--', label='Train/Test Split')
plt.legend()
plt.grid()
plt.show()
```

- **Facebook prophet Model**

```python
import pandas as pd
import matplotlib.pyplot as plt
from prophet import Prophet
from sklearn.metrics import r2_score

# Prepare the training data for Prophet
train_data_prophet = train_data.copy()
train_data_prophet.reset_index(inplace=True)  # Reset index to have a column for date
train_data_prophet.columns = ['ds', 'y']      # Prophet requires 'ds' for date and 'y' for target variable

# Initialize and fit the Prophet model
model = Prophet(yearly_seasonality=True)      # Enable yearly seasonality
model.fit(train_data_prophet)

# Make predictions on the test data
future = model.make_future_dataframe(periods=len(test_data), freq='M')
forecast = model.predict(future)

# Extract only the predicted values for the test period
predictions = forecast.set_index('ds').loc[test_data.index, 'yhat']

# Calculate error metrics
mae = r2_score(test_data['count'], predictions)
mse = r2_score(test_data['count'], predictions)

print(f"R Squared: {mae}")
print(f"R Squared: {mse}")

# Optional: Plot the predictions vs actual values
plt.figure(figsize=(10, 6))
plt.plot(train_data.index, train_data['count'], label='Train', marker='o')
plt.plot(test_data.index, test_data['count'], label='Test', marker='o')
plt.plot(predictions.index, predictions, label='Predictions', color='g', marker='o')
plt.xlabel('Date')
plt.ylabel('Count')
plt.title('Prophet Model Predictions')
plt.axvline(test_data.index[0], color='r', linestyle='--', label='Train/Test Split')
plt.legend()
plt.grid()
plt.show()
```

```python
from prophet import Prophet

# Initialize and fit the Prophet model with additional parameters
model = Prophet(
    yearly_seasonality=True,
    weekly_seasonality=False,        # Disable weekly seasonality if not needed
    daily_seasonality=False,         # Disable daily seasonality if not needed
    seasonality_mode='multiplicative',  # Change seasonality mode
    changepoint_prior_scale=0.1,        # Increase changepoint flexibility
    n_changepoints=30                  # Increase the number of changepoints
)


# Fit the model
model.fit(train_data_prophet)

# Make predictions on the test data
future = model.make_future_dataframe(periods=len(test_data), freq='M')
forecast = model.predict(future)

# Extract only the predicted values for the test period
predictions = forecast.set_index('ds').loc[test_data.index, 'yhat']

# Calculate error metrics
r2 = r2_score(test_data['count'], predictions)

print(f"R Squared: {r2}")

# Plot the predictions vs actual values
plt.figure(figsize=(10, 6))
plt.plot(train_data.index, train_data['count'], label='Train', marker='o')
plt.plot(test_data.index, test_data['count'], label='Test', marker='o')
plt.plot(predictions.index, predictions, label='Predictions', color='g', marker='o')
plt.xlabel('Date')
plt.ylabel('Count')
plt.title('Prophet Model Predictions')
plt.axvline(test_data.index[0], color='r', linestyle='--', label='Train/Test Split')
plt.legend()
plt.grid()
plt.show()
```

```python
from prophet import Prophet
from prophet.diagnostics import cross_validation, performance_metrics
import pandas as pd
import matplotlib.pyplot as plt

# Initialize the Prophet model with additional parameters
model = Prophet(
    yearly_seasonality=True,
    weekly_seasonality=False,
    daily_seasonality=False,
    seasonality_mode='multiplicative',
    changepoint_prior_scale=0.1,  # Adjust as necessary
    n_changepoints=30
)

# Add custom 6-month seasonality (biannual)
model.add_seasonality(name='biannual', period=182.5, fourier_order=5)

# Fit the model on training data
model.fit(train_data_prophet)

# Make predictions on the test data
future = model.make_future_dataframe(periods=len(test_data), freq='M')
forecast = model.predict(future)

# Extract only the predicted values for the test period
predictions = forecast.set_index('ds').loc[test_data.index, 'yhat']

# Calculate R squared and additional metrics for evaluation
r2 = r2_score(test_data['count'], predictions)
print(f"R Squared: {r2}")


# Plot the predictions vs actual values
plt.figure(figsize=(12, 6))
plt.plot(train_data.index, train_data['count'], label='Train', marker='o')
plt.plot(test_data.index, test_data['count'], label='Test', marker='o')
plt.plot(predictions.index, predictions, label='Predictions', color='g', marker='o')
plt.xlabel('Date')
plt.ylabel('Count')
plt.title('Enhanced Prophet Model Predictions with 6-Month Seasonality')
plt.axvline(test_data.index[0], color='r', linestyle='--', label='Train/Test Split')
plt.legend()
plt.grid()
plt.show()
```

```python
from prophet.diagnostics import cross_validation, performance_metrics

# Set shorter durations for cross-validation due to limited data
df_cv = cross_validation(
    model, 
    initial='540 days',  # About 18 months (1.5 years), enough to cover some initial data
    period='90 days',    # 3-month gap between cutoffs
    horizon='180 days'   # 6-month forecast window
)

# Calculate performance metrics
df_p = performance_metrics(df_cv)

# Display cross-validation metrics without 'r2'
print("Cross-Validation Metrics:")
print(df_p[['horizon', 'mape', 'rmse', 'mse', 'mae']].head())

```
